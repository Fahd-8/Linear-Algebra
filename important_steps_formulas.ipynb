## *Important SVD Formulas:*

---

#### 1. The Main Formula:
> **A = U × Σ × V^T**

---

#### 2. Finding V (Right Singular Vectors):
> **A^T × A → find eigenvectors → that's V**

---

#### 3. Finding U (Left Singular Vectors):
> **A × A^T → find eigenvectors → that's U**

---

#### 4. Finding Singular Values:
> **σ = √λ**

Where λ = eigenvalues of A^T × A

---

#### 5. Finding Eigenvalues:
> **det(A - λI) = 0**

---

#### 6. Reconstruction with k components:
> **A ≈ σ1×U1×V1^T + σ2×U2×V2^T + ... + σk×Uk×Vk^T**

---

#### In Order of Calculation:

```
1. Calculate A^T × A
        ↓
2. det(A^T × A - λI) = 0  → eigenvalues (λ)
        ↓
3. σ = √λ  → singular values
        ↓
4. eigenvectors of A^T × A → V
        ↓
5. eigenvectors of A × A^T → U
        ↓
6. A = U × Σ × V^T  ✅
```
