```python
# The Idea:

# Use SVD to compress an image by keeping only the most important singular values
```

```python
# Why images?

# An image is just a matrix of numbers (pixel values)
# SVD finds the most important patterns in those pixels
# Keep top patterns в†’ smaller file, similar quality
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1 - Create a simple image (matrix of numbers)
# WHY: Image is just a matrix вЂ” perfect for SVD
image = np.array([
    [100, 100, 100, 200, 200, 200],
    [100, 100, 100, 200, 200, 200],
    [100, 100, 100, 200, 200, 200],
    [200, 200, 200, 100, 100, 100],
    [200, 200, 200, 100, 100, 100],
    [200, 200, 200, 100, 100, 100],
])

print(f"Original image matrix:\n{image}")
print(f"Original size: {image.shape}")
```

```python

# Step 2 - Apply SVD
# WHY: Break image into U, ОЈ, V^T
U , sigma, VT = np.linalg.svd(image)
print(f"\nSingular values: {sigma}")
print("в†‘ Bigger = more important pattern")
```

```python
print(f"\nLeft Singular values: {U}")
```

```python
print(f"\nRight Singular values: {VT}")
```

```python
# Step 3 - Reconstruct with only TOP 1 singular value
# WHY: Keep only most important pattern вЂ” throw rest away
k = 1
compressed = sigma[0] * np.outer(U[:, 0], VT[0, :])
print(f"\nCompressed image (using only top {k} singular value):")
print(np.round(compressed))
```

```python
# Step 4 - Check how close it is to original
error = np.round(np.linalg.norm(image - compressed), 2)
print(f"\nError from original: {error}")
print("Lower error = better compression ")
```

```python
# Step 5 - Try with top 2 singular values
k = 2
compressed2 = sum(sigma[i] * np.outer(U[:, i], VT[i, :]) for i in range(k))
error2 = np.round(np.linalg.norm(image - compressed2), 2)
print(f"\nWith top 2 singular values вЂ” Error: {error2}")
print("Even closer to original! ")
```

```python
# Original matrix
#       в†“
# np.linalg.svd() в†’ U, ОЈ, V^T
#       в†“
# Choose k (how many patterns to keep)
#       в†“
# Reconstruct with only top k
#       в†“
# Compressed version 
```

```python
import numpy as np

# Original matrix (4x5) вЂ” imagine this is any dataset
A = np.array([
    [1,  2,  3,  4,  5],
    [2,  4,  6,  8,  10],
    [3,  6,  9,  12, 15],
    [4,  8,  12, 16, 20]
])
```

```python
print(f"Original matrix ({A.shape[0]}x{A.shape[1]}):")
print(A)
```

```python
# Step 1 - Apply SVD
U, sigma, VT = np.linalg.svd(A)
print(f"\nSingular values: {np.round(sigma, 2)}")
print("в†‘ See how first value is huge, rest are basically zero!")
```

```python
# Step 2 - Keep only top k=1
k = 1
compressed = sum(sigma[i] * np.outer(U[:, i], VT[i, :]) for i in range(k))
print(f"\nCompressed matrix (k={k}):")
print(np.round(compressed, 1))
```

```python
# Step 3 - Check error
error = np.round(np.linalg.norm(A - compressed), 4)
print(f"\nError: {error}")
```

```python
# Step 4 - How much data did we save?
original_size = A.shape[0] * A.shape[1]
compressed_size = k * (A.shape[0] + A.shape[1] + 1)
print(f"\nOriginal size: {original_size} numbers")
print(f"Compressed size: {compressed_size} numbers")
print(f"Saved: {original_size - compressed_size} numbers! ")
```
